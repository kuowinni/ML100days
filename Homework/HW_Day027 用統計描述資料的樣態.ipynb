```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
## 輸入資料
boy=[164,175,183,173,176,159,165,165,169,151
      ,156,163,169,144,170,177,165,160,164,171]

girl=[169,170,162,154,183,173,169,167,170,185,162,
     175,168,151,181,170,182,156,159,160]
```
